# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data

```python
#Importing Libraries
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
###Connecting to web driver######
driver = webdriver.Chrome('chromedriver.exe')
######opening website Naukari.com on automated chrome window####
url = "https://www.naukri.com/"
driver.get(url)
```

```python
#Finding web element for search bar
search_job = driver.find_element(By.CLASS_NAME,"suggestor-input")
search_job.send_keys("Data Analyst")

###Finding web element for Location as it is showing the same class we are using absolute XPATH#####
search_location = driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')
search_location.send_keys("Banglore")
#####For search button#####
search_button = driver.find_element(By.CLASS_NAME,"qsbSubmit")
search_button.click()
```

```python
##Creating empty list for Data scraping job title####

job_titles = []
title_tags = driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title_tags:
    job_titles.append(i.text)

job_titles    
```

```python
#Creating empty list for scraping location####
Location = []
loc_tags = driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 locWdth"]')
for i in loc_tags:
    Location.append(i.text)

Location    
```

```python
##Creating empty list to scrape data for company names###
company=[]
comp_tags= driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in comp_tags:
    company.append(i.text)
    
company    
```

```python
#Creating empty list for Scraping data of Experience
Experience=[]
exp_tags = driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 expwdth"]')
for i in exp_tags:
    Experience.append(i.text)
    
Experience    
```

```python
#First 10 Jobs data
df = pd.DataFrame({'Job_title':job_titles,'Location':Location,'Company':company,'Experience':Experience})
df.head(10)
```

```python
##Closing the automated window#####
driver.close()
```

# Q11: Write s python program to display list of 50 Most expensive cars in the world (i.e. Company name, Model name and Price) from https://www.motor1.com/
This task will be done in following steps:
1. First get the webpage https://www.motor1.com/
2. Then You have to click on the List option from Dropdown menu on left side.
3. Then click on 50 most expensive cars in the world..
4. Then scrap the mentioned data and make the

```python
#Importing necessary Libraries

import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
###connecting to webdriver###
driver = webdriver.Chrome('chromedriver.exe')
url =" https://www.motor1.com/"
driver.get(url)
```

```python
###Clicking on Dropdown menu#####
list_click = driver.find_element(By.CLASS_NAME,"m1-hamburger-button")
list_click.click()
```

```python
####Clicking on List#####
lists_click = driver.find_element(By.XPATH,'/html/body/div[3]/div[1]/div[3]/ul/li[4]/a')
lists_click.click()
```

```python
####Clicking on most expensive cars#####
cars_click = driver.find_element(By.XPATH,'/html/body/div[2]/div[9]/div[1]/div[1]/div/div/div[1]/div/div[1]/h3/a')
cars_click.click()
```

```python
###Scraping list of cars####
Cars=[]
scrapped_cars= driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in scrapped_cars:
    Cars.append(i.text)

    
    
```

```python
##Scraping Price####
Price=[]
scrapped_price = driver.find_elements(By.XPATH,'//div[@class="pre-center padding b-padding-skin pre-center-feature"]//p//strong')
for i in scrapped_price:
    Price.append(i.text.split(',')[0])
    
####Indexing#####
indexed_cars=Cars[:50]
indexed_price = Price[:50]
####Data Frame######
df = pd.DataFrame({'Cars & CO.':indexed_cars,'Price':indexed_price})
df
  
    
```
